```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#dist plot allow us to show univariate observation distribution
#histogram
sns.distplot(tips['total_bill'],kde= False,bins= 30)
```

```python
#bivariate graphs
#scatter plot
sns.jointplot(x= 'total_bill',y='tip',data=tips,kind='kde')
#kind = 'reg'/'hex'
```

```python
# plots graphs for all numerical variable
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
#################### CATEGORICAL PLOTS  ####################
```

```python
tips.head()
```

```python
import numpy as np
```

```python
#barplot
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
# its showing average value by default
#in estimator we can mention our own function
```

```python
sns.countplot(x='sex',data=tips)
```

```python
# boxplot
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')

# add another category variable in hue argument to bifurcate plot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#scatter plot based on category
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
#combining scatter plot and strip plot with violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')

#dont use for large datasets
```

```python
#create all kind of plot using factor plot, mention plot name in kind arg.
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')

```

```python
##################  MATRIX PLOT #######################
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
tc=tips.corr()
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='black',linewidths=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
################### REGRESSION PLOTS ###################
```

```python
#linear model plot
sns.lmplot(x='total_bill',y='tip',data =tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
#scatter_kws to change size of markers

```

```python
sns.lmplot(x='total_bill',y='tip',data =tips,col='sex',row='time')
#aspect and size arg for changing size of plots
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# to create customise plots for all variable
#sns.set_style('whitegrid')
g=sns.PairGrid(iris,despine=False)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
#g.map(sns.distplot,'total_bill')
g.map(plt.scatter,'total_bill','tip')
```

```python
sns.boxplot(x='class',y='age',data=td,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=td)
```

```python
sns.countplot('sex',data=td)
```

```python
tdcr=td.corr()
sns.heatmap(tdcr,cmap='coolwarm')
plt.title('titanic corr plot')
```
